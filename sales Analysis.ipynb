## what was the best month for sales nd how much earned that month?

### Add new columns

```python
df_new['Month']=df_new['Order Date'].str[0:2]
```

```python
df_new.head()
```

```python
type('Month')
```

```python
df_new['Month']=df_new['Month'].astype(int)
```

## find 'or' and delete it

```python
temp_df = df_new[df_new['Month'].str[0:2] == 'Or']
temp_df.head()
```

```python
df_new = df_new[df_new['Month'].str[0:2] != 'Or']
```

```python
df_new['Month']=df_new['Month'].astype('int32')
```

```python
df_new.head()
```

```python
df_new['Quantity Ordered']=pd.to_numeric(df_new['Quantity Ordered'])# to convert string into int
df_new['Price Each']=pd.to_numeric(df_new['Price Each'])# to convert string into float
```

```python
df_new['Sales']= df_new['Quantity Ordered']*df_new['Price Each']
```

```python
df_new.head()
```

```python
df_new.groupby('Month').Sales.sum().nlargest()
```

## Which City sold the most product

```python
df_new['City']=df_new['Purchase Address'].apply(lambda x: x.split(',')[1])
df_new.head()
```

```python
df_new['City']=df_new['Purchase Address'].apply(lambda x: x.split(',')[1]+' '+x.split(',')[2])
```

```python
df_new.head()
```

```python
df_new.groupby('City').Sales.sum().nlargest()
```
